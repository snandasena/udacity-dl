```python
# Imports 
import matplotlib
# %matplotlib inline
```

```python
# Imports
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
# Load the image
image = mpimg.imread("../images/exit-ramp.jpg")
```

```python
plt.imshow(image)
plt.show()
```

```python
# import opencv
import cv2
```

```python
# convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, cmap="gray")
plt.show()
```

```python
#define a kernal size for Guassian smothing/ blurring
kernal_size = 3
blur_gray = cv2.GaussianBlur(gray, (kernal_size, kernal_size), 0)
```

```python
# define parameters for Canny and run it
low_threashole = 50
high_threashold = 150
edges = cv2.Canny(blur_gray, low_threashole, high_threashold)
```

```python
plt.imshow(edges, cmap="Greys_r")
plt.show()
```
