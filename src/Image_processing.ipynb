```python
# Imports
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
# Load the image
image = mpimg.imread("../images/test.jpg")
```

```python
# Show the loaded image
plt.imshow(image)
plt.show()
```

```python
image.shape
```

```python
ysize = image.shape[0]
xsize = image.shape[1]

# Copy an image for further modifications
color_select = np.copy(image)
```

```python
ysize , xsize
```

```python
# Tune to colour parameters
red_threshold = 200
green_threshold = 200
blue_threashold = 200

rgb_threashold = [red_threshold, green_threshold, blue_threashold]
```

```python
# pixels below the thresholds
thresholds = (image[:,:,0] < rgb_threashold[0]) \
            | (image[:,:,1] < rgb_threashold[1]) \
            | (image[:,:,2] < rgb_threashold[2])
color_select[thresholds] = [0,0,0]
```

```python
plt.imshow(color_select)
plt.show()
```
